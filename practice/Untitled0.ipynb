```python
age = int(input('나이를 입력하세요: '))

if age>19:
    print('성인입니다')
elif age>14:
    print('청소년입니다')
elif age>6:
    print('초딩입니다')
elif age>0:
    print('애기입니다')
else:
    print('커신입니다')
print(f'입력한 나이: {age}')
```

```python
id = input('아이디를 입력하세요: ')

if (length := len(id))<3:
    print(f'너무 짧아. (입력한 글자 수: {length})')
else:
    print(f"{id} 사용가능")
```
